## 1. Pengantar Data Science

### 1.1 Definisi Data Science

**Data Science** adalah bidang interdisipliner yang menggabungkan ilmu statistika, matematika, pemrograman, dan domain pengetahuan bisnis untuk mengekstrak wawasan bermakna dari data. Data Science memungkinkan organisasi untuk membuat keputusan berbasis data (*data-driven decision making*).

Menurut **Provost & Fawcett (2013)**, Data Science adalah seperangkat prinsip fundamental yang memandu ekstraksi pengetahuan dan wawasan dari data. Proses ini mencakup pengumpulan, pembersihan, eksplorasi, pemodelan, hingga komunikasi hasil analisis.

### 1.2 Lifecycle Data Science

```
┌──────────────────────────────────────────────────────────┐
│              SIKLUS HIDUP DATA SCIENCE                   │
│                                                          │
│  1. Business Understanding → Apa masalah yang ingin     │
│                               diselesaikan?             │
│                                                          │
│  2. Data Acquisition       → Dari mana data didapat?    │
│                                                          │
│  3. Data Preprocessing     → Membersihkan & merapikan   │
│                               data                      │
│                                                          │
│  4. Exploratory Analysis   → Memahami pola & distribusi │
│                                                          │
│  5. Modeling (opsional)    → Membuat prediksi/klasifik.  │
│                                                          │
│  6. Evaluation & Deploy    → Mengomunikasikan insight   │
└──────────────────────────────────────────────────────────┘
```

### 1.3 Peran Data Science dalam E-Business

Dalam konteks **e-business** (bisnis elektronik), Data Science berperan untuk:

| Peran | Contoh Aplikasi |
|---|---|
| Analisis penjualan | Memahami tren pendapatan bulanan |
| Segmentasi pasar | Mengelompokkan pelanggan berdasarkan perilaku |
| Optimasi logistik | Menganalisis pola pengiriman & lokasi tujuan |
| Pemahaman produk | Mengetahui kategori produk terlaris |
| Analisis pembayaran | Memahami preferensi metode bayar pelanggan |

## 3. Load Dataset

````python
import sys
import os
sys.path.insert(0, os.path.join(os.getcwd(), '..', 'src'))

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

from utils import load_all_excel, set_style
set_style()

print('✅ Library berhasil diimport')
````

````python
# Load semua file Excel dari folder CLEAN
CLEAN_FOLDER = os.path.join('..', 'temp_data', 'Clean_Dataset', 'CLEAN')

df_raw = load_all_excel(CLEAN_FOLDER)
print(f'\nShape data: {df_raw.shape}')
print(f'Kolom: {list(df_raw.columns)}')
````

### 3.2 Struktur Dataset

````python
print('=== INFORMASI DATASET ===')
print(f'Jumlah baris     : {len(df_raw):,}')
print(f'Jumlah kolom     : {len(df_raw.columns)}')
print(f'Jumlah file sumber: {df_raw["_source_file"].nunique()}')
print(f'\nDaftar kolom:')
for i, col in enumerate(df_raw.columns, 1):
    print(f'  {i:2d}. {col}')
````

````python
# Tipe data setiap kolom
print('=== TIPE DATA ===')
print(df_raw.dtypes.to_string())
````

### 3.3 Distribusi Data per File Sumber (per Bulan)

````python
# Distribusi jumlah baris per file (per bulan)
per_file = df_raw.groupby('_source_file').size().sort_index()
print('=== JUMLAH TRANSAKSI PER BULAN ===')
for f, n in per_file.items():
    label = f.replace('_clean.xlsx', '').replace('Sales', ' ')
    print(f'  {label:<30} : {n:,} transaksi')
print(f'\n  TOTAL                          : {per_file.sum():,} transaksi')
````

````python
# Visualisasi distribusi per bulan
fig, ax = plt.subplots(figsize=(14, 5))

labels = [f.replace('_clean.xlsx','').replace('Sales',' ') for f in per_file.index]
ax.bar(labels, per_file.values, color='#2a9d8f', alpha=0.85, edgecolor='white')
ax.set_title('Jumlah Transaksi per Bulan', fontsize=14, fontweight='bold')
ax.set_ylabel('Jumlah Transaksi')
ax.set_xlabel('Bulan')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
os.makedirs('../reports/figures', exist_ok=True)
plt.savefig('../reports/figures/01_transaksi_per_bulan.png', dpi=150, bbox_inches='tight')
plt.show()
print('✅ Figure disimpan')
````

### 3.4 Simpan Data Gabungan

````python
# Simpan ke data/processed/
os.makedirs('../data/processed', exist_ok=True)
df_raw.to_csv('../data/processed/all_months_clean.csv', index=False)
print('✅ Data berhasil disimpan ke: data/processed/all_months_clean.csv')
print(f'   Ukuran file: {os.path.getsize("../data/processed/all_months_clean.csv") / 1024:.1f} KB')
````
